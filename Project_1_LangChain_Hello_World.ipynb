### Importing Langchain Google GenAI Package

```python
import langchain_google_genai as genai
```

### Setting up Google API Key

```python
from google.colab import userdata
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
```

### Setting up Google Gemini LLM Model

```python
model : genai.ChatGoogleGenerativeAI = genai.ChatGoogleGenerativeAI(
    model = "gemini-2.0-flash-exp",
    api_key=GOOGLE_API_KEY,
)
```

### Interacting with Google Gemini LLM model through Google API

```python
response1 = model.invoke("Who is national poet of Pakistan")
```

```python
print(response1.content)
```
